2.Explain the purpose of comments in HTML and provide an example of how to use comments in an HTML documents

Ans: Comments in HTML are used to add notes or explanations in the code. They are not visible on the web page and help in understanding and debugging the code.

```python
<!-- This is a comment -->
<!DOCTYPE html>
<html>
<head>
    <title>Comments Example</title>
</head>
<body>

<h1>Hello World</h1>
<!-- This heading is displayed on the page -->

</body>
</html>
```

3) Write an HTML program that includes a heading, a paragraph of text, a horizontal line, and a line break.
Arrange these elements to create a simple web page layout

```python
<!DOCTYPE html>
<html>
<head>
    <title>Simple Web Page</title>
</head>
<body>

<h1>Welcome to My Page</h1>

<p>This is a simple paragraph of text.</p>

<hr>

<p>This is another line after a horizontal line.<br>
This text appears after a line break.</p>

</body>
</html>
```
